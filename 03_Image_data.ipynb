# 이미지 흐리게 하기

```python
image = cv2.imread("/content/drive/MyDrive/MachineLearning_class/고우주-1일차_실습코드/01_Data Structures/images/cybertruck.jpg", cv2.IMREAD_GRAYSCALE)

# 이미지를 흐리게
image_blurry = cv2.blur(image, (5, 5))

plt.imshow(image_blurry, cmap = "gray"), plt.axis("off")
plt.show()
```

```python
image_very_blurry = cv2.blur(image, (100, 100))

plt.imshow(image_very_blurry, cmap = "gray"), plt.xticks([])
plt.show()
```

```python
kernel = np.ones((5, 5)) / 25.0
kernel
```

```python
image_kernel = cv2.filter2D(image, -1, kernel)

plt.imshow(image_kernel, cmap = "gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

# 이미지를 선명하게

```python
image = cv2.imread("/content/drive/MyDrive/MachineLearning_class/고우주-1일차_실습코드/01_Data Structures/images/cybertruck.jpg", cv2.IMREAD_GRAYSCALE)

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# 선명하게
image_sharp = cv2.filter2D(image, -1, kernel)

plt.imshow(image_sharp, cmap = "gray"), plt.axis("off")
plt.show()
```

# 평균 색을 특성으로 인코딩

```python
image_bgr = cv2.imread("/content/drive/MyDrive/MachineLearning_class/고우주-1일차_실습코드/01_Data Structures/images/plane_256x256.jpg", cv2.IMREAD_COLOR)

# 각 채널의 평균을 계산
channels = cv2.mean(image_bgr)

observation = np.array([(channels[2], channels[1], channels[0])])
observation
```
